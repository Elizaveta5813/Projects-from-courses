```python
from sklearn.datasets import load_digits
import numpy as np
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
```

```python
digits = load_digits()
```

```python
X = digits['data']
y = digits['target']
```

```python
X_test = X[int(0.75 * X.shape[0]):,:]
X_train = X[:int(0.75 * X.shape[0]),:]

y_test = y[int(0.75 * X.shape[0]):]
y_train = y[:int(0.75 * X.shape[0])]
```

```python
def metric(x, y):
    
    return (np.sum((x - y)** 2)) 
```

```python
y_prediction = []

for value in X_test:
    
    index_min = 0
    value_min = metric(value, X_train[0])
    
    for ind, train_value in enumerate(X_train):
        
        if (metric(value, train_value) < value_min):
            
            index_min = ind
            value_min = metric(value, train_value)
            
    y_prediction.append(y_train[index_min])
```

```python
accuracy_score(y_test, y_prediction)
```

```python
err = 1 - accuracy_score(y_test, y_prediction)

with open("accuracy.txt", 'w') as fout:
    
    fout.write(str(err))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
res = RandomForestClassifier(n_estimators = 1000)
res.fit(X_train, y_train)
```

```python
accuracy_score(y_test, res.predict(X_test))
```

```python
with open('accuracy_forest.txt', 'w') as fout:
    
    fout.write(str(1 - accuracy_score(y_test, res.predict(X_test))))
```
